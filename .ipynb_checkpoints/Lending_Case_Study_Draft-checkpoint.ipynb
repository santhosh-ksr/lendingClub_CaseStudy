```python
import pandas as pd
import numpy as np
import seaborn as sns


import warnings
warnings.filterwarnings('ignore')
```

# IMPORTING & READING THE DATASET


```python
dataframe=pd.read_csv("loan.csv")
dataframe.head()
```

# NUMBER OF ROWS AND COLUMNS IN DATASET


```python
rows, Cols = dataframe.shape
print(f"Dataset has {rows} rows and {Cols} columns")
```

# Visualizing the feature based on Loan Status

```python
#removing white spaces from the Columns
dataframe = dataframe.rename(columns= lambda x: x.strip(),inplace=False)
dataframe.describe()
```

```python
#Removing all null columns
dataframe.dropna(how='all',axis=1,inplace=True)
rows,cols = dataframe.shape
print(f"After removing all null columns, there are {rows} rows and {cols} Columns")
```

```python
#Identifying Percentage of missing values from each Columns
missing_values = (dataframe.isnull().sum()* 100 /len(dataframe))
missing_values.sort_values(ascending = False)
```

```python
#Removing Columns with more than 60% of missing values(from the above they are next_pymnt_d, mths_since_last_record, mths_since_last_delinq)
#dataframe.describe().columns

dataframe.drop(['next_pymnt_d','mths_since_last_record','mths_since_last_delinq'],axis=1,inplace = True)
rows,cols = dataframe.shape
print(f"After removing Cols with > 60% of missing values, now dataset has {rows} rows and {cols} Columns")
```

```python
dataframe = dataframe.drop_duplicates()
rows,columns = dataframe.shape
print(f"There are {rows} after deleting duplicate rows")
```

```python
#Chekcing if there are any duplicate columns, if exists would be removed
df = dataframe.loc[:,dataframe.columns.duplicated()].copy()
rows,cols = df.shape
print(f"There are {cols} duplicate Columns")
```

```python
dataframe.info()
```

```python
#Dropping few columns which are not required for analysis
dataframe.drop(['member_id','url','desc','zip_code'],axis=1,inplace=True)
dataframe.shape
```

```python
#Now dropping customer behaviour variables, which would not be available during the prediction for customer loan approval process
behaviour_var=['delinq_2yrs',
               'earliest_cr_line',
               'inq_last_6mths',
               'open_acc',
               'pub_rec',
               'revol_bal',
               'revol_util',
               'total_acc',
               'initial_list_status',
               'out_prncp',
               'out_prncp_inv',
               'total_pymnt',
               'total_pymnt_inv',
               'total_rec_prncp',
               'total_rec_int',
               'total_rec_late_fee',
               'collection_recovery_fee',
               'last_pymnt_d',
               'last_pymnt_amnt',
               'last_credit_pull_d',
               'collections_12_mths_ex_med',
               'chargeoff_within_12_mths',
               'tax_liens']
behaviour_var
```

```python
dataframe.drop(behaviour_var, axis = 1,inplace = True)
rows,cols = dataframe.shape
print(f"After removing the behaviour variables, the dataset has {rows} rows and {cols} columns")
```

```python
missingValues = dataframe.isnull().sum()* 100/len(dataframe)
missingValues[missingValues > 0]
```

The above mentioned values mentioned are in Percentage

```python
#1. Ignoring rows what has null values for title as it is very low
#2. Filling emp_length with the mode value(the value that occurs most times)
#3. Filling pub_rec_bankruptcies with the mean value
#4. Filling emp_title with the unknown values in place of Null


dataframe = dataframe[~(dataframe.title.isnull())]
dataframe['emp_length']=dataframe['emp_length'].fillna(dataframe['emp_length'].mode()[0])
dataframe['pub_rec_bankruptcies']=dataframe['pub_rec_bankruptcies'].fillna(dataframe['pub_rec_bankruptcies'].mean())
dataframe['emp_title'].fillna('Unknown',inplace= True)
```

```python
#Deriving new Columns from date field

dataframe['issue_d'] = pd.to_datetime(dataframe['issue_d'],format='%b-%y')
dataframe['issue_d']
dataframe['issue_month']=dataframe['issue_d'].dt.month
dataframe['issue_year']=dataframe['issue_d'].dt.year
dataframe[['issue_year','issue_month','issue_d']].head()
```

```python
#Verifying dataTtypes for each column and will change if required
dataframe.dtypes
```

```python
# The following data types changes will be made 
#      1. Will change term to integer by stripping months
#      2. Will change interest rate to float
#      3. will change emp_lenth to int
```

```python
#change term to integer by stripping months
dataframe.term.value_counts(normalize=True)
dataframe.term = dataframe.term.replace('months','',regex=True).astype(int)
dataframe.term.dtype
```

```python
# change interest rate to float by removing percentage
dataframe['int_rate'] = dataframe['int_rate'].str.strip().str.strip('%').astype('float')
dataframe.int_rate.dtype
```

```python
#Change emp_length to int
import re
dataframe['emp_length'] = dataframe['emp_length'].apply(lambda x:re.findall('\d+',str(x))[0])
dataframe['emp_length'] = dataframe['emp_length'].apply(lambda x: pd.to_numeric(x))
dataframe.emp_length.dtype
```

```python
dataframe.int_rate.head()
```

```python
dataframe.loan_status.value_counts(normalize='True')*100
```

From the above, ~83% of loans are Fully Paid, and 14% are charged_off/defaulted and 3% are in Progress, 
will remove in-progress loan status as they will not help in finding the recommendations

```python
#sns.set_theme(style='whitegrid')
dataframe = dataframe[~(dataframe['loan_status']=='Current')]
dataframe.loan_status.value_counts(normalize='True')
```

```python
##Set values for 'loan_status' variable's unqiue entires. Assign '0' to 'fully paid' and set 'charged off' to '1' (as our target is determine 'defaulter' factors)
dataframe["loan_status"] = dataframe["loan_status"].replace('Fully Paid','0')
dataframe["loan_status"] = dataframe["loan_status"].replace('Charged Off','1') #Set Target Variable
dataframe.head()
```

```python
#Analysis using term of months on the Loan status
data=pd.crosstab(dataframe['term'],dataframe['loan_status'])
data.div(data.sum(1).astype('float'),axis=0).plot(kind='bar', stacked=True, color=['orange', 'lightgreen'])
plt.title('Term Vs Loan Status')
plt.xlabel('term')
plt.show()
```

Above analysis shows the risk of getting defaulters is those who opt for long term(i.e. 60Months)

```python
#Analysis using term of Int Rate Vs Loan status
sns.boxplot(dataframe['int_rate'],dataframe['loan_status'])
plt.show()
```

Loan with highest int_rate are more likely to charge-off/Defaulted

```python
#Analysis on which grade people are like to be defaulters
data=pd.crosstab(dataframe['grade'],dataframe['loan_status'])
data.div(data.sum(1).astype('float'),axis=0).plot(kind='bar', stacked=True, color=['lightgreen', 'orange'])
plt.title('Grade Vs Loan Status')
plt.xlabel('Grade')
plt.show()
```

From the above analysis, F & G graded people are high risk of defaulters
